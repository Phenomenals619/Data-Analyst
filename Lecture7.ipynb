```python
import pandas as pd
data = pd.read_excel("empolyee_Performance _Data.xlsx")
```

```python
data.head(10)
data.tail(10)
```

```python
data.info()
```

```python
mean_value=data[["Q1 Performance (%)","Q2 Performance (%)","Q3 Performance (%)","Q4 Performance (%)"]].mean()
print(mean_value)
```

```python
mean_value
```

```python
median_value=data[["Q1 Performance (%)","Q2 Performance (%)","Q3 Performance (%)","Q4 Performance (%)"]].median()
median_value
```

```python
mode_value=data[["Q1 Performance (%)","Q2 Performance (%)","Q3 Performance (%)","Q4 Performance (%)"]].mode()
mode_value
```

```python
quarterly_avg = data.mean(numeric_only=True)
plt.figure(figsize=(8, 5))
sns.barplot(x=quarterly_avg.index, y=quarterly_avg.values, palette="viridis")
plt.title("Average Perfomance Across Quarters")
plt.ylabel("Perfomance (%)")
plt.xlabel("Quarters")                                                                                                             
plt.show()
```

```python
df_melted = data.melt(id_vars=["Employee Name"], var_name="Quarter", value_name="Performance")

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_melted, x="Quarter", y="Performance", hue="Employee Name", marker="o")
plt.title("Performance Trends of Employee Across Quarters")
plt.ylabel("Performance (%)")
plt.xlabel("Quarters")
plt.legend(title="Employee Name")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxenplot(data=df_melted, x="Quarter", y="Performance", palette="Set2")
plt.title("Performance Distribution Across Quarters")
plt.ylabel("Performance (%)")
plt.xlabel("Quarters")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(data.set_index("Employee Name").T, annot=True, cmap="coolwarm", fmt=".0f")
plt.title("Employee Performance Heatmap")
plt.ylabel("Quarters")
plt.xlabel("Employees")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random as r
```

```python
file_path="Supermart Grocery Sales - Retail Analytics Dataset.csv"
df=pd.read_csv(file_path)
print("Basic Info: ")
print(df.info())
print("\nSummary Statistics")
print(df.describe())
```

```python
print(df.head())
df['Total']=df[['Sales','Discount','Profit']].sum(axis=1)
```

```python
city_sales=df.groupby('City')['Sales'].sum().sort_values(ascending=False)
print(city_sales)
```

```python
plt.figure(figsize=(15,10))
city_sales.plot(kind='bar',color='red')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
import random 
df['Gender']=random.choices(['Male','Female'],weights=[0.5,0.5],k=len(df))
```

```python
gender_count=df['Gender'].value_counts()
print(gender_count)
```

```python
plt.figure(figsize=(5,5))
gender_count.plot(kind='pie',autopct='%1.1f%%', colors=['gray', 'pink'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```
